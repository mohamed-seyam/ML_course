```python
import numpy as np
import idx2numpy
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X_train = np.array(idx2numpy.convert_from_file("data/train-images.idx3-ubyte"))
Y_train = np.array(idx2numpy.convert_from_file("data/train-labels.idx1-ubyte"))
X_test  = np.array(idx2numpy.convert_from_file("data/t10k-images.idx3-ubyte"))
Y_test  = np.array(idx2numpy.convert_from_file("data/t10k-labels.idx1-ubyte"))
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1]*X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1]*X_test.shape[2])
X_train.shape
```

```python
for elem in range(0,len(Y_train)) : 
    if Y_train[elem] != 3 :
        Y_train[elem] = 0

for elem in range(0,len(Y_test)) : 
    if Y_test[elem] != 3 : 
        Y_test[elem] = 0
```

```python
model = LogisticRegression(max_iter = 100)
model.fit(X_train,Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
for pred in predictions :
    print(pred)
```

```python
_, axes = plt.subplots(nrows=1, ncols=30,figsize=(10, 3))
for ax, image, prediction in zip(axes, X_test, predictions):
    ax.set_axis_off()
    image = image.reshape(28, 28)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f' {prediction}')
```
